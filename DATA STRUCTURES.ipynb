```python
def process_a(lines):
    result = []
    
    for line in lines :
        if len(line) > 1:
            result.append(last_word(line))
            
    return result
process_a(lines)
```

```python
def process_b(lines):
    result = []
    
    for line in lines :
        if len(line) > 1:
            result.append(line[0])
            
    return result

process_b(lines)
```

```python
def process(f, lines):
    result = []
    
    for line in lines:
        if len(line) > 1:
            result.append(f(line))
            
    return result

process(str.upper, lines)
```
